```python
intgrad_scores = pd.read_csv("../Results/IntegratedGradient/integrated_gradients.csv", header=None)
```

```python
max_scores = intgrad_scores.max()
min_scores = intgrad_scores.min()
mean_scores = intgrad_scores.mean()
stddev_scores = intgrad_scores.std()
AE = max_scores - min_scores
```

```python
RE = np.abs(AE)
# fig, ax = plt.subplots(1, 2)
# sns.distplot(max_scores, hist=False)
# sns.distplot(min_scores, hist=False)
# sns.distplot(mean_scores, hist=False)
# sns.distplot(max_scores, hist=False)
sns.distplot(RE, hist=False)
plt.xlabel("Absolute Error")
plt.ylabel("Density")
# sns.distplot(RE, hist=False, ax=ax[1])
plt.show()
```
